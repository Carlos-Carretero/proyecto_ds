# Fundamentos COMPLETOS de DataFrames en Pandas

```python
import pandas as pd
import numpy as np
```

## 1. Creación básica de DataFrame

```python
data = {
    'Nombre': ['Ana', 'Luis', 'Marta'],
    'Edad': [23, 35, 29],
    'Ciudad': ['Madrid', 'Barcelona', 'Valencia']
}
df = pd.DataFrame(data)
df
```

## 2. Selección y filtrado

```python
df['Nombre']
df[df['Edad'] > 25]
```

## 3. Modificación

```python
df['Salario'] = [50000, 60000, 55000]
df = df.drop(columns=['Ciudad'])
df
```

## 4. Formas de crear DataFrames

```python
# Lista de listas
data = [
    ['Ana', 23, 'Madrid'],
    ['Luis', 35, 'Barcelona']
]
pd.DataFrame(data, columns=['Nombre','Edad','Ciudad'])
```

```python
# Lista de diccionarios
data = [
    {'Nombre':'Ana','Edad':23},
    {'Nombre':'Luis','Edad':35}
]
pd.DataFrame(data)
```

```python
# Desde NumPy
data = np.array([
    ['Ana',23,'Madrid'],
    ['Luis',35,'Barcelona']
])
pd.DataFrame(data, columns=['Nombre','Edad','Ciudad'])
```

```python
# Desde Series
nombres = pd.Series(['Ana','Luis'])
edades = pd.Series([23,35])
pd.DataFrame({'Nombre':nombres,'Edad':edades})
```

## 5. Índices

```python
df = pd.DataFrame({
    'Nombre':['Ana','Juan','Pedro'],
    'Edad':[23,25,30]
})
df.set_index('Nombre', inplace=True)
df
```

```python
# Reset index
df.reset_index(inplace=True)
df
```

```python
arrays = [['Ana','Ana','Luis','Luis'],['Math','Science','Math','Science']]
index = pd.MultiIndex.from_arrays(arrays, names=('Nombre','Asignatura'))
df_multi = pd.DataFrame({'Notas':[90,85,88,92]}, index=index)
df_multi
```

```python
data = {
    'Nombre':['Ana','Luis','Ana'],
    'Edad':[23,35,22],
    'Ciudad':['Madrid','Barcelona','Madrid']
}
df_group = pd.DataFrame(data)
df_group.groupby('Ciudad').mean(numeric_only=True)
```

## 8. loc e iloc

```python
df = pd.DataFrame({'Nombre':['Ana','Luis','Marta'],'Edad':[23,35,29]})

df.iloc[0]
df.iloc[0:2]
df.iloc[0,1]
df.iloc[0:2,0:2]
```

```python
df.loc[0]
df.loc[df['Edad'] > 25]
df.loc[(df['Edad'] > 25) & (df['Nombre'] == 'Luis')]
```
