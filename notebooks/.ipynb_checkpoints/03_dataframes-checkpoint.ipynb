# Fundamentos de DataFrames en Pandas

## 1. Creación de DataFrame

```
import pandas as pd

# Crear un DataFrame desde un diccionario
data = {
    'Nombre': ['Ana', 'Luis', 'Marta'],
    'Edad': [23, 35, 29],
    'Ciudad': ['Madrid', 'Barcelona', 'Valencia']
}

df = pd.DataFrame(data)
df
```

```
# Este DataFrame almacena información de personas
# Cada columna representa una variable y cada fila un registro
```

## 2. Selección y filtrado

```
df['Nombre']
```

```
# Filtrar personas mayores de 25 años
df[df['Edad'] > 25]
```

## 3. Modificación de datos

```
# Agregar una columna
df['Salario'] = [50000, 60000, 55000]

# Eliminar columna
df = df.drop(columns=['Ciudad'])

df
```

## 4. Otras formas de crear DataFrames

```
# Desde lista
data = [
    ['Ana', 23],
    ['Luis', 35]
]

df2 = pd.DataFrame(data, columns=['Nombre', 'Edad'])
df2
```

## 5. Atributos

```
df.shape
df.columns
df.dtypes
df.head()
```

```
# shape muestra filas y columnas
# columns muestra los nombres de las columnas
```

## 6. Métodos básicos

```
df['Edad'].mean()
df['Edad'].max()
df['Edad'].min()
df['Edad'].sum()
```

## 7. loc e iloc

```
# Acceso por posición
df.iloc[0]

# Acceso por etiqueta
df.loc[0]
```

```
# Filtrar con loc
df.loc[df['Edad'] > 25]
```

## Reflexión

El uso de DataFrames en Pandas permite organizar los datos de forma estructurada en filas y columnas, facilitando su manipulación y análisis.

Gracias a sus múltiples funciones, es posible filtrar, agrupar, transformar y analizar grandes volúmenes de información de manera eficiente. Esto mejora la comprensión de los datos, ya que permite visualizar patrones, realizar cálculos rápidamente y obtener resultados claros.

En un proyecto de análisis de datos, los DataFrames son fundamentales porque optimizan el tiempo de trabajo y permiten realizar operaciones complejas con pocas líneas de código.
