```python
import pandas as pd
import numpy as np
```

```python
# 1
s1 = pd.Series(range(1, 11))
print(s1)

# 2
materias = pd.Series(["Programación", "Matemáticas", "Bases de Datos", "IA", "Estadística"])
print(materias)

# 3
precios = pd.Series([2500, 3700, 5900, 1200, 4800])
print(precios.dtype)

# 4
booleanos = pd.Series([True, False, True, True, False])
print(booleanos)

# 5
serie = pd.Series([5,10,15,20,25,30])
print(serie.head(3))
```

```python
# 1
s = pd.Series({'a':10,'b':20,'c':30})
print(s)

# 2
notas = pd.Series({
    'Matemáticas':4.5,
    'Inglés':3.8,
    'Programación':5.0
})
print(notas)

# 3
temperaturas = pd.Series({
    'Lunes':30,
    'Martes':28,
    'Miércoles':27
})
print(temperaturas.index)

# 4
edades = pd.Series({
    'Ana':20,
    'Luis':22,
    'María':19
})
print(edades.max())

# 5
edades['Carlos'] = 23
print(edades)
```

```python
# 1
arr = np.random.randint(0,100,10)
serie = pd.Series(arr)
print(serie)

# 2
serie = pd.Series(np.arange(1,11))
print(serie)

# 3
decimales = pd.Series(np.array([1.1,2.2,3.3,4.4]))
print(decimales)

# 4
celsius = pd.Series([20,25,30,35])
fahrenheit = celsius * 9/5 + 32
print(fahrenheit)

# 5
arr = np.random.randint(1,10,5)
serie = pd.Series(arr) * 10
print(serie)
```

```python
# 1
s = pd.Series([100]*5)
print(s)

# 2
s = pd.Series(["Aprendiendo Pandas"]*8)
print(s)

# 3
s = pd.Series([True]*10)
print(s)

# 4
s = pd.Series([3.14]*5, dtype='float64')
print(s)

# 5
a = pd.Series([1,2,3,4,5])
b = pd.Series([10]*5)
print(a+b)
```

```python
# 1
s = pd.Series(['Juan','Ana','Luis'], index=['ID1','ID2','ID3'])
print(s)

# 2
paises = pd.Series(['Colombia','México','Argentina'], index=['CO','MX','AR'])
print(paises)

# 3
notas = pd.Series([4.2,3.8,4.9], index=['Pedro','Ana','Luis'])
print(notas)

# 4
serie = pd.Series([10,20,30,40])
serie.index = ['a','b','c','d']
print(serie)

# 5
print(serie['b'])
```

```python
# 1
s = pd.Series([1,'texto',np.nan])
print(s)

# 2
print(s.dtype)

# 3
s = pd.Series([1,'dos',3.0,True])
print(s)

# 4
s = pd.Series([pd.Timestamp("2025-01-01"),10,"hola"])
print(s)

# 5
print(s.astype(str))
```

```python
# 1
s = pd.Series([1,2,3,4])
print(s.index)
print(s.values)
print(s.dtype)

# 2
s = pd.Series(range(20))
print(s.size)
print(s.shape)
print(s.empty)

# 3
s = pd.Series([1,2,2,3])
print(s.is_unique)

# 4
s = pd.Series([1,2,np.nan])
print(s.hasnans)

# 5
s = pd.Series(["texto largo ejemplo"]*5)
print(s.nbytes)
```

```python
# 1
s = pd.Series(range(10))
print(s.iloc[2])

# 2
s = pd.Series(range(10), index=list("abcdefghij"))
print(s.loc['d'])

# 3
print(s.iloc[:3])

# 4
print(s.loc[['b','d','f']])

# 5
print(s.iloc[2])
print(s.loc['c'])
```

```python
s1 = pd.Series([1,2,3])
s2 = pd.Series([4,5,6])

print(s1+s2)
print(s2-s1)
print(s1*2)
print(s2/2)
print((s1+s2)/2)
```

```python
s = pd.Series([1,2,3,4])

print(s.apply(lambda x: x**2))

nombres = pd.Series(["ana","juan","luis"])
print(nombres.map(str.upper))

print(s.apply(np.sqrt))

def clasificar(x):
    return "alto" if x>2 else "bajo"

print(s.apply(clasificar))

reemplazo = {"ana":"ANA","juan":"JUAN","luis":"LUIS"}
print(nombres.map(reemplazo))
```

```python
s = pd.Series([5,1,3,2,4])

print(s.sort_values())
print(pd.Series(["c","a","b"]).sort_values())
print(s.sort_index())

print(s.nlargest(3))
print(s.nsmallest(3))
```

```python
s = pd.Series([10,40,60,80,30])

print(s[s>50])
print(s[s%2==0])

texto = pd.Series(["python","pandas","sql"])
print(texto[texto.str.contains("pan")])

print(s[(s>30) & (s<70)])

print(s[s>30].index)
```

```python
s = pd.Series([1,2,np.nan,4])

print(s.fillna(0))
print(s.dropna())
print(s.isna().sum())
print(s.fillna(s.mean()))
print(s.fillna(100))
```

```python
s1 = pd.Series([1,2,3], index=['a','b','c'])
s2 = pd.Series([4,5,6], index=['b','c','d'])

print(s1+s2)
print(s1.add(s2, fill_value=0))
print(s1.combine_first(s2))
```

```python
s = pd.Series([10,20,30,40], index=['a','b','c','d'])

print(s.reindex(['a','b','c','d','e']))
print(s.reindex(['d','c','b','a']))
print(s.reindex(['a','b','c','d','e'], fill_value=0))
```
